## Задача 
Создать модель мемристора на основе оксида гафния. Сравнить характеристики модели с реальным устройством

В модели главными параметрами являются: концентрация носителей: weak points, и распределение температуры по оксиду. Данные параметры будут высчитываться дифференциальных уравнений для концетрации и температуры (соответсвенно уравнения диффузии и теплопроводности 

По распределению концентрации расчитываем проводимость и ток через оксид

Распределение концентрации в зависимости от напряжения на электроде
