## Тема НИР:
Создание полной цифровой модели мемристора на основе оксида HfO2.

## Актуальность НИР:
С развитием технологий и увеличением потребностей в энергоэффективных и высокопроизводительных устройствах, мемристоры становятся все более актуальными. Для создания мемристоров на основе новых материалов необходимо создания его полной модели на основе уже изученных оксидов. Оксид HfO2 является одним из наиболее исследованных материалов. Создание полной цифровой модели мемристора на основе HfO2 с использованием программного обеспечения Synopsys TCAD позволит более точно прогнозировать поведение будущих мемристоров.

## Цель НИР: 
Описание полноценной модели мемристора на основе HfO2 c использованием программного обеспечения Synopsys TCAD, которая позволит более точно описывать RRAM на основе новых материалов.

## Задачи:
1) Исследование физических и электрических свойств оксида HfO2.
2) Создание математической модели, описывающей динамику работы мемристора.
3) Реализация модели в среде Synopsys TCAD для проведения симуляций и получения данных.
4) Сравнние полученных результатов с эксперементальными данными и валидация модели мемристора.
